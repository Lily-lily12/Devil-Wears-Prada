#RANDOM FOREST CLASSIFIER

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Sample data (replace this with reading the actual dataset)
df=pd.read_csv('clothes.csv')

# Drop the 'Description' column as it's not needed for the model
df = df.drop('Description', axis=1)

# Separate features and target variable
x = df.drop('Dress Name', axis=1)
y = df['Dress Name']

# One-hot encoding for the categorical features
x_encoded = pd.get_dummies(x)

# Label encoding for the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2, random_state=42)

# Train the Random Forest classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(x_train, y_train)

# Predict on the test set
y_pred = classifier.predict(x_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python

# Function to predict dress name from input string
def predict_dress(input_str):
    # Split the input string into a list
    input_list = input_str.split(',')
    # Create a DataFrame from the input list
    input_df = pd.DataFrame([input_list], columns=x.columns)
    # One-hot encode the input DataFrame
    input_encoded = pd.get_dummies(input_df)
    # Align the input DataFrame with the training DataFrame (handle missing columns)
    input_encoded = input_encoded.reindex(columns=x_encoded.columns, fill_value=0)
    # Predict the dress name
    pred = classifier.predict(input_encoded)
    # Decode the prediction to the original label
    pred_label = label_encoder.inverse_transform(pred)
    return pred_label[0]

# Example usage
x1 = "Sunny,Bright,Floral,Casual,Loose"
result = predict_dress(x1)
print("Predicted Dress Name:", result)
```

# ANN MODEL

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf

# Load the dataset
df = pd.read_csv('clothes.csv')

# Drop the 'Description' column as it's not needed for the model
df = df.drop('Description', axis=1)

# Separate features and target variable
x = df.drop('Dress Name', axis=1)
y = df['Dress Name']

# One-hot encoding for the categorical features
x_encoded = pd.get_dummies(x)

# Label encoding for the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Convert the input data to float32
x_encoded = x_encoded.astype('float32')

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2, random_state=42)

# Building the ANN
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=len(set(y_encoded)), activation='softmax'))

# Compiling the ANN
ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training the ANN on the Training set
ann.fit(x_train, y_train, batch_size=32, epochs=100)

# Function to predict dress name from input string
def predict_dress(input_str):
    # Split the input string into a list
    input_list = input_str.split(',')
    # Create a DataFrame from the input list
    input_df = pd.DataFrame([input_list], columns=x.columns)
    # One-hot encode the input DataFrame
    input_encoded = pd.get_dummies(input_df)
    # Align the input DataFrame with the training DataFrame (handle missing columns)
    input_encoded = input_encoded.reindex(columns=x_encoded.columns, fill_value=0)
    # Convert the input data to float32
    input_encoded = input_encoded.astype('float32')
    # Predict the dress name
    pred = ann.predict(input_encoded)
    # Decode the prediction to the original label
    pred_label = label_encoder.inverse_transform([pred.argmax()])[0]
    return pred_label

# Example usage
x1 = "Sunny,Bright,Floral,Casual,Loose"
result = predict_dress(x1)
print("Predicted Dress Name:", result)
```
